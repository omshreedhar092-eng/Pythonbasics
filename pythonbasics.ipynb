14.describe a scenario where a while loop is more suitable than a for loop .

Scenario: Validating User Input

Consider a program that asks a user to enter their age. The program needs to keep asking indefinitely until the user enters a valid, positive integer.

Why while is better here:

You do not know ahead of time how many attempts the user will take to enter a correct age. They might enter valid input on the first try, or the tenth.
The while loop naturally fits this structure because it continues as long as the input is invalid.

Why a for loop is unsuitable here:

A for loop requires a defined sequence to iterate over. You cannot define a sequence that says "loop an indeterminate number of times until something is true."

You could technically force a for loop to work using an infinite loop (for _ in iter(int, 1): or similar workarounds with break), but this is an unnatural and less readable use of the loop structure.

```python
'8.write a program to demonstrate the use of logiocal operators.'
'Program: Demonstrating Logical Operators'

' Define variables representing a user's status'
user_age = 25
is_student = True
has_coupon = False

print(f"User Age: {user_age}")
print(f"Is Student: {is_student}")
print(f"Has Coupon: {has_coupon}\n")

'--- 1. The 'and' operator ---'
'Both conditions must be True for the result to be True.'
print("--- Using 'and' operator ---")
if user_age >= 18 and is_student == True:
    print("Result: Eligible for student discount.")
else:
    print("Result: Not eligible for student discount (must be 18+ AND a student).")


' --- 2. The 'or' operator ---'
'At least one condition must be True for the result to be True.'
print("\n--- Using 'or' operator ---")
if is_student == True or has_coupon == True:
    print("Result: Eligible for some kind of promotion.")
else:
    print("Result: Not eligible for any promotion.")


'--- 3. The 'not' operator ---'
'This reverses the Boolean result of the condition that follows it.'
print("\n--- Using 'not' operator ---")
if not has_coupon:
    # This evaluates to "If NOT False" which is True
    print("Result: The user does NOT have a coupon.")
else:
    print("Result: The user HAS a coupon.")


' --- 4. Combining all operators ---'
' A complex condition for VIP eligibility'
is_vip = False
has_purchased_recently = True

print("\n--- Complex Condition (Eligibility Check) ---")
if (user_age >= 30 or is_vip) and has_purchased_recently and not is_student:
    print("Result: User is eligible for the VIP loyalty program.")
else:
    print("Result: User does not meet all complex eligibility criteria.")
```
